```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True, cmap='rainbow', fmt='.2f',vmin=-1,vmax=1,center=0)
plt.title('Correlation Heatmap')
print("Inference:")
print("1. There is a high positive correlation (0.7) between Average_Score and Reviewer_Score. This indicates if a hotel has a higher average score from reviews, the new reviewers also tend to give it a high rating.")  

print("2. There is a moderate negative correlation (-0.3) between Negative_Review word count and Reviewer_Score. So hotels with higher word count in negative reviews tend to have lower reviewer ratings.")
```

```python
worst_hotels = df.sort_values('Reviewer_Score')[:10]

hotel_names = worst_hotels['Hotel_Name'].values
scores = worst_hotels['Reviewer_Score'].values

plt.figure(figsize=(8, 4))
sns.barplot(x=scores, y=hotel_names, palette='rocket')

plt.xlabel('Reviewer Score')
plt.ylabel('Hotel Name')
plt.title('10 Worst Hotels')  
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
best_hotels = df.sort_values('Reviewer_Score', ascending=False).head(10)  

hotel_names = best_hotels['Hotel_Name'].values
scores = best_hotels['Reviewer_Score'].values

plt.figure(figsize=(8, 4)) 
sns.barplot(x=scores, y=hotel_names, palette='crest')

plt.xlabel('Reviewer Score')
plt.ylabel('Hotel Name')
plt.title('10 Best Hotels')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```
